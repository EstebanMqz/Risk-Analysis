$W(t) = \int_0^t \xi(s) \, ds$
$W(t)$ is the Wiener process (Brownian motion) at time $t$.<br>
$\int_0^t \xi(s) \, ds$ represents the integral of the Gaussian white noise.<br><br>
$xi(s)$ is the Gaussian white noise process.<br>
$s$ representing time.<br>
$ds$ differential of time.<br>
$\xi(s)$ is a stochastic process with mean zero and delta correlation with uncorrelated values. 

$$W(t) = \int_0^t \xi(s)\,ds$$

where $\xi(s)\,ds \equiv dW(s)$ represents the infinitesimal increment of the Wiener process. 

Since $W(0) = 0$, we have:

$$\int_0^t \xi(s)\,ds = W(t) - W(0) = W(t)$$
